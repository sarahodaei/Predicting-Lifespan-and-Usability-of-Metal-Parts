```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
#encoding partType
encoded_data = df.copy()
oneHotEncoder = OneHotEncoder()
onehot_enc = oneHotEncoder.fit(encoded_data[['partType']])
print('The categories are: ', onehot_enc.categories_)
OHT = onehot_enc.transform(encoded_data[['partType']]).toarray()
encoded_data['partType_0'] = OHT[:,0]
encoded_data['partType_1'] = OHT[:,1]
encoded_data['partType_2'] = OHT[:,2]
encoded_data['partType_3'] = OHT[:,3]

#microstructure
onehot_enc = oneHotEncoder.fit(encoded_data[['microstructure']])
print('The categories are: ', onehot_enc.categories_)
OHT = onehot_enc.transform(encoded_data[['microstructure']]).toarray()
encoded_data['microstructure_0'] = OHT[:,0]
encoded_data['microstructure_1'] = OHT[:,1]
encoded_data['microstructure_2'] = OHT[:,2]

#castType
onehot_enc = oneHotEncoder.fit(encoded_data[['castType']])
print('The categories are: ', onehot_enc.categories_)
OHT = onehot_enc.transform(encoded_data[['castType']]).toarray()
encoded_data['castType_0'] = OHT[:,0]
encoded_data['castType_1'] = OHT[:,1]
encoded_data['castType_2'] = OHT[:,2]

#seedLocation
label_encoder = LabelEncoder()
encoded_data['seedLocation_encoded'] = label_encoder.fit_transform(encoded_data['seedLocation'])
#all categorical features are encoded in encoded_data_pre + numericals before scaling
encoded_data_pre = encoded_data.drop(['partType', 'microstructure', 'castType', 'seedLocation'], axis=1)
display(encoded_data_pre)

from sklearn.model_selection import train_test_split

X_non_test, X_test, y_non_test, y_test = train_test_split(encoded_data_pre, encoded_data_pre['Lifespan'], test_size=0.20, shuffle=True, random_state=0)

X_train, X_valid, y_train, y_valid = train_test_split(X_non_test, y_non_test, test_size=0.25, shuffle=True, random_state=0)

print("Training set shape:", X_train.shape, y_train.shape)
print("Validation set shape:", X_valid.shape, y_valid.shape)
print("Testing set shape:", X_test.shape, y_test.shape)

from sklearn.preprocessing import StandardScaler
numerical_features = ['coolingRate', 'quenchTime', 'forgeTime', 'HeatTreatTime',
                      'Nickel%', 'Iron%', 'Cobalt%', 'Chromium%', 'smallDefects',
                      'largeDefects', 'sliverDefects']
categorical_features = ['partType_0', 'partType_1', 'partType_2', 'partType_3',
                       'microstructure_0', 'microstructure_1', 'microstructure_2',
                       'castType_0', 'castType_1', 'castType_2',
                       'seedLocation_encoded']
#training data
X_train_num = X_train[numerical_features]
X_train_cat = X_train[categorical_features]
scaler = StandardScaler()
X_train_num_scaled = scaler.fit_transform(X_train_num)
#scaled numbers and encoded together
X_train_scaled = pd.concat([pd.DataFrame(X_train_num_scaled, columns=numerical_features, index=X_train.index), X_train_cat], axis=1)
#display(X_train_scaled)

#validation data
X_valid_num = X_valid[numerical_features]
X_valid_num_scaled = scaler.transform(X_valid_num)
X_valid_scaled = pd.concat([pd.DataFrame(X_valid_num_scaled, columns=numerical_features, index=X_valid.index), X_valid[categorical_features]], axis=1)

#test data
X_test_num = X_test[numerical_features]
X_test_num_scaled = scaler.transform(X_test_num)
X_test_scaled = pd.concat([pd.DataFrame(X_test_num_scaled, columns=numerical_features, index=X_test.index), X_test[categorical_features]], axis=1)
# Display the scaled validation and test data (optional)
#display(X_valid_scaled)
#display(X_test_scaled)
```

```python
from scipy.stats import f_oneway
categorical_features_encoded = ['partType_0', 'partType_1', 'partType_2', 'partType_3',
                               'microstructure_0', 'microstructure_1', 'microstructure_2',
                               'castType_0', 'castType_1', 'castType_2',
                               'seedLocation_encoded']
numerical_feature = 'Lifespan'

for categorical_feature in categorical_features_encoded:
    groups = [encoded_data_pre[encoded_data_pre[categorical_feature] == category][numerical_feature]
              for category in encoded_data_pre[categorical_feature].unique()]
    f_stat, p_value = f_oneway(*groups)

    print(f"ANOVA results for {categorical_feature} vs. {numerical_feature}:")
    print(f"F-statistic: {f_stat}")
    print(f"P-value: {p_value}")
    print("-" * 20)
```

```python
#visualizing the Anova-test Results
import matplotlib.pyplot as plt

f_statistics = []
p_values = []
features = categorical_features_encoded

for categorical_feature in features:
    groups = [
        encoded_data_pre[encoded_data_pre[categorical_feature] == category][numerical_feature]
        for category in encoded_data_pre[categorical_feature].unique()
    ]
    f_stat, p_value = f_oneway(*groups)
    f_statistics.append(f_stat)
    p_values.append(p_value)

significance_level = 0.05

plt.figure(figsize=(10, 6))
bars = plt.bar(features, f_statistics, color=['red' if p < significance_level else 'blue' for p in p_values])
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)

plt.title("ANOVA F-statistic for Categorical Features vs Lifespan")
plt.xlabel("Categorical Features")
plt.ylabel("F-statistic")
plt.xticks(rotation=45, ha='right')

for bar, p_value in zip(bars, p_values):
    height = bar.get_height()
    if p_value < significance_level:
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'p={p_value:.3f}', ha='center', va='bottom', color='red')
    else:
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'p={p_value:.3f}', ha='center', va='bottom', color='black')

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
import numpy as np

# Fit the best model on the training data
best_model.fit(X_train, y_train)

# Make predictions on the test set
y_test_pred = best_model.predict(X_test)

# Evaluate the model on the test data
accuracy_test = accuracy_score(y_test, y_test_pred)
precision_test = precision_score(y_test, y_test_pred, average='weighted')
recall_test = recall_score(y_test, y_test_pred, average='weighted')
f1_test = f1_score(y_test, y_test_pred, average='weighted')

# Display classification metrics
print("Test Set Evaluation:")
print(f"Accuracy: {accuracy_test:.4f}")
print(f"Precision: {precision_test:.4f}")
print(f"Recall: {recall_test:.4f}")
print(f"F1 Score: {f1_test:.4f}")

# Generate a detailed classification report
print("\nClassification Report:")
print(classification_report(y_test, y_test_pred))

# Plot Learning Curve
def plot_learning_curve(estimator, X, y, cv=None, scoring='accuracy', n_jobs=None, train_sizes=np.linspace(0.1, 1.0, 5)):
    train_sizes, train_scores, test_scores = learning_curve(estimator, X, y, cv=cv, scoring=scoring, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.figure(figsize=(10, 6))
    plt.plot(train_sizes, train_scores_mean, 'o-', label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', label="Cross-validation score")
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.1)
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std, test_scores_mean + test_scores_std, alpha=0.1)
    plt.title("Learning Curve")
    plt.xlabel("Training Size")
    plt.ylabel("Score")
    plt.legend(loc="best")
    plt.grid()
    plt.show()

# Plot Confusion Matrix
def plot_confusion_matrix(y_true, y_pred, classes):
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)
    plt.figure(figsize=(8, 6))
    disp.plot(cmap="Blues", values_format="d")
    plt.title("Confusion Matrix")
    plt.show()

# Plot the learning curve
plot_learning_curve(best_model, X_train, y_train, cv=5)

# Plot the confusion matrix
plot_confusion_matrix(y_test, y_test_pred, classes=np.unique(y_test))
```
